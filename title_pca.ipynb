```python
import pandas as pd
import numpy as np 
from sklearn.preprocessing import StandardScaler
```

```python
df1 = pd.read_csv(r'C:/Users/johd003/Documents/GitHub/data/title_stats.csv')
df2 = pd.read_csv(r'C:/Users/johd003/Documents/GitHub/data/title_labor_type_pivot.csv').fillna(0)
df = pd.merge(df1, df2, how = "inner", on = "Title")
df.head()
```

```python
y = df.Title.values
x_df = df.iloc[:,1:df.shape[1]]
x_df.drop(x_df.columns[2], axis=1, inplace=True)
x = x_df.values
# Standardizing the features
#x = StandardScaler().fit_transform(x)
```

```python
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='median')
x = imp.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```
